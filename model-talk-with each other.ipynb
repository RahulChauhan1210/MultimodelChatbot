```python
deepseek_model = "deepseek-r1:1.5b"
llama_model = "llama3.2"

# Initialize the conversation
deepseek_message = "Hello, who are you?"
conversation_rounds = 5

for round in range(conversation_rounds):
  
    # DeepSeek talks to Llama
    print(f"DeepSeek: {deepseek_message}")
    llama_response = ollama.chat(
        model=llama_model,
        messages=[{'role': 'user', 'content': deepseek_message}],
         options={"num_predict": 500}
    )
    llama_message = llama_response['message']['content']
    print(f"Llama: {llama_message}")

    # Llama responds back to DeepSeek
    deepseek_response = ollama.chat(
        model=deepseek_model,
        messages=[{'role': 'user', 'content': llama_message}],
         options={"num_predict": 500}
    )
    deepseek_message = deepseek_response['message']['content']
```
